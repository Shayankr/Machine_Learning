## Deep Learning Frameworks

Deep learning frameworks can be considered as a library or tools which helps to built a deep learning models more easily and quickly.

These deep learning libraries help in implementing artificial neural networks.

Top Deep Learning Frameworks

1. TensorFlow

   * Tensor Board -- used for visualizing TensorFlow computations and graphs.
   * TensorFlow Serving -- used for rapid deployment of new algorithm/experiments while retaining the same server architecture and APIs.
  
  
2. Keras-- High level API which can run on top of TensorFlow, Theano and CNTK.

    Model building is easy, stacking layers on top of each-other.
    * Recurrent Neural Network
    * Convolutional Neural Network


3. PyTorch -- scientific computing framework developed by "facebook".
    * Pythonic in nature.
    * dynamic Computational graph.


4. Sonnet
5. MXNet -- developed by "Apache Software Foundation".
    * R
    * Python
    * Scala
    * Julia

6. Swift for TensorFlow
7. Gluon
8. DL4J -- Deep Learning programming library written for "Java".
    * Hadoop
    * Spark
9. ONNX
10. Chainer
